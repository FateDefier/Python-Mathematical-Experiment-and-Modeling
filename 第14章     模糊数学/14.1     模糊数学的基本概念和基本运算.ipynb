```python
# 模糊关系的合成运算
import numpy as np

a = np.array([0.3, 0.35, 0.1])  # (3,),一维数组
aa = np.tile(a, (len(a), 1))  # 广播成(3, 3)

b = np.array([[0.3, 0.5, 0.2], [0.2, 0.2, 0.4], [0.3, 0.4, 0.2]])  # (3, 3)

c = np.minimum(aa.T, b)  # 求两个矩阵对应元素的最小值,注意“转置”,R的第一行与S的第一列比较,(3, 3)

T = c.max(axis=0)  # 逐列取最大值,(3)

print(f"合成运算的结果: T={T}")
```
