### 1. 择近原则

例 14.7 设标准库集合 $ A_1 = (0.4, 0.3, 0.5, 0.3) $, $ A_2 = (0.3, 0.3, 0.4, 0.4) $, $ A_3 = (0.2, 0.3, 0.3, 0.3) $, 待识别集合 $ B = (0.2, 0.3, 0.4, 0.3) $, 试确定 $ B $ 属于哪一类.

解 计算欧几里得贴近度分别为
$$ N(B, A_1) = 0.8882, \quad N(B, A_2) = 0.9293, \quad N(B, A_3) = 0.95, $$

由此集合 $ B $ 与集合库中 $ A_3 $ 最为接近, 因此 $ B $ 归于 $ A_3 $ 类.

```python
import numpy as np

A = np.array([[0.4, 0.3, 0.5, 0.3], [0.3, 0.3, 0.4, 0.4], [0.2, 0.3, 0.3, 0.3]])

B = np.array([0.2, 0.3, 0.4, 0.3])

N = []

for i in range(len(A)):
    N.append(1 - (np.linalg.norm(A[i] - B)) / 2)
    
print(f"各欧几里得贴进度分别为: {N}")
```

### 2. 最大隶属原则

例 14.8 考虑人的年龄问题, 分为年轻、中年、老年三类, 分别对应三个 $ F $ 集 $ A_1, A_2, A_3 $. 设论域 $ U = (0, 100] $, 且对 $ x \in (0, 100] $, 有

$$
A_1(x) = 
\begin{cases} 
1, & 0 < x \leqslant 20, \\
1 - 2 \left( \dfrac{x - 20}{20} \right)^2, & 20 < x \leqslant 30, \\
2 \left( \dfrac{x - 40}{20} \right)^2, & 30 < x \leqslant 40, \\
0, & 40 < x \leqslant 100; \\
1 - 2 \left( \dfrac{ \ \cdot \ \cdot \ }{20} \right)^2, & 60 < x \leqslant 70, \\
1, & 70 < x \leqslant 100; 
\end{cases}
$$

$$
A_2(x) = 1 - A_1(x) - A_3(x) = 
\begin{cases} 
0, & 0 < x \leqslant 20, \\
2 \left( \dfrac{x - 20}{20} \right)^2, & 20 < x \leqslant 30, \\
1 - 2 \left( \dfrac{x - 40}{20} \right)^2, & 30 < x \leqslant 40, \\
1, & 40 < x \leqslant 50, \\
1 - 2 \left( \dfrac{x - 50}{20} \right), & 50 < x \leqslant 60, \\
2 \left( \dfrac{x - 70}{20} \right)^2, & 60 < x \leqslant 70, \\
0, & 70 < x \leqslant 100. 
\end{cases}
$$

某人 40 岁, 根据上式, $ A_1(40) = 0 $, $ A_2(40) = 1 $, $ A_3(40) = 0 $, 则

$$
A_2(40) = \max \{ A_1(40), A_2(40), A_3(40) \} = 1,
$$

按最大隶属原则, 他应该是中年人.

又当 $ x = 35 $ 时, $ A_1(35) = 0.125 $, $ A_2(35) = 0.875 $, $ A_3(35) = 0 $. 可见 35 岁的人应该是中年人.
