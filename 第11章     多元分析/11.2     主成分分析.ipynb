```sklearn.decomposition``` 模块的 ```PCA``` 函数实现主成分分析，调用格式如下：

```sklearn.decomposition.PCA(n_components=None, copy=True)```（decomposition分解 PCA即Principal Components Analysis主成分分析）

其中，
n_components: 类型为 int 或字符串  

缺省时默认为 None，所有成分被保留  
赋值为 int，比如 n_components=2,提取两个主成分  
赋值为(0, 1)上的浮点数，自动选择主成分的个数，满足信息贡献率的要求（svd_solver='full）

copy：类型为 bool，即 True 或 False表示是否在运行算法时，将原始数据复制一份  

缺省时默认为True，表示运行PCA算法后，原始训练数据的值不会有任何改变，因为是在原始数据的副本上进行运算  

若为False，表示运行PCA算法后，原始训练数据的值会改变，因为是在原始数据上进行降维计算



````python
import numpy as np
from sklearn.decomposition import PCA
# from sklearn.preprocessing import StandardScaler  # 标准化scaler比例尺，缩放

a = np.loadtxt("Pdata11_7.txt")
b = np.r_[a[:, 1:4], a[:, -3:]]  # 构造数据矩阵
scaler = StandardScaler()
b_scaled = scaler.fit_transform(b)

pca = PCA()  # 构造模型
md = pca.fit(b)  # 训练模型

print(f"特征值为: {md.explained_variance_}")
print(f"各主成分的贡献率: {md.explained_variance_ratio_}")
print(f"奇异值为: {md.singular_values_}")
print(f"各主成分的系数为:\n{md.components_}")  # 每行是一个主成分的系数
"""
下面直接计算特征值和特征向量, 和PCA对比
"""
covariance_matrix = np.cov(b.T)  # 协方差矩阵
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)  # 求特征值和特征向量

print(f"特征值: {eigenvalues}")
print(f"特征向量:\n{eigenvectors}")  # 每一列的特征向量变成了主成分的一行的各个系数（行向量）
print(f"各主成分的贡献率为: {eigenvalues / np.sum(eigenvalues)}")
````

注 11.1   
(1) 从上面程序运行结果可以看出， PCA 函数使用**协方差矩阵**作的主成分分析. 主成分分析也可以使用**相关系数矩阵**，两者计算结果略有差异，使用相关系数阵作主成分分析，相当千对数据进行了标准化处理．  

(2) 从程序的运行结果看，主成分的系数可以相差一个负号，因为特征向量乘以－1 后仍然为特征向量．用主成分分析做评价等其他模型时，觉得很难把握，主成分的系数变成了相反数，模型的解释都变了．

第一主成分 $F_1$ 是身高值 $x_1$、胸围值 $x_2$ 和体重值 $x_3$ 的加权和，当一个学生的 $F_1$值较大时， 可以推断他较高或较胖或又高又胖；反之，当一个学生比较魁梧时，所对应的尺值一般也较大， 故第一主成分是反映学生身材是否魁梧的综合指标，一般称之为 “大小” 因子. 而在第二主成分的表达式中， 身高$x_1$ 前的系数为正， 而胸围 $x_2$ 和体重 $x_3$ 前的系数为负， 当一个学生的$F_1$值较大时，说明 $x_1$ 的值较大， 而$x_2$ 和 $x_3$ 的值相对较小，即该生较高且瘦；反之， 瘦高型学生的 $F_2$ 值会较大， 故$F_2$是反映学生体型特征的综合指标，一般称之为＂形状”因子．

````python
import numpy as np
from scipy.stats import zscore

a = np.loadtxt("Pdata11_8.txt")
b = np.delete(a, 0, axis=1)  # 删除第一列数据
c = zscore(b)  # z-score也叫标准分数，用来标准化
r = np.corrcoef(c.T)  # 计算相关系数矩阵，correlation coeffitient相关系数corrcorf
d, e = np.linalg.eig(r)  # 计算特征值和特征向量
rate = d / d.sum()  # 计算各主成分的贡献率

print(f"特征值为: {d}")
print(f"特征向量为:\n{e}")
print(f"各主成分的贡献率为: {rate}")

k = 1  # 提出主成分的个数，根据各主成分的得分率取得
F = e[:, :k]  # 第1列（索引为0）所有行，即第一列全部元素，特征向量u_1
score_mat = c.dot(F)  # 计算主成分得分矩阵，向量点积
score1 = score_mat.dot(rate[0:k])  # 计算各评价对象的得分,rate[0:k]=0.92200854
score2 = - score1  # 通过观测，调整得分的正负号

print(f"各评价对象的的得分为:\n{score2}")

# array.argsort()返回排序后各元素在原数组的原始索引（类似秩统计量）+1 确保索引从1开始
index = score1.argsort() + 1  # 排序后的每个元素在原数组中的位置

print(f"从高到低各个城市的编号顺序为: {index}")
````
