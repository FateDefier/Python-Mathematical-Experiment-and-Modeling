# Pandas 数据处理

***数据清洗*** 与 ***预处理*** 常见步骤：

1. ***缺失值处理*** ：识别并填补缺失值，或删除含缺失值的行/列

2. ***重复数据处理*** ：检查并删除重复数据，确保每条数据唯一

3. ***异常值处理*** ：识别并处理异常值。如极端值、错误值

4. ***数据格式转换*** ：转换数据类型或进行单位转换，如日期格式转换

5. ***标准化与归一化*** ：对数值型数据进行标准化(`sklearn.process.Z-score`)或归一化(`Min-Max`)

6. ***类别数据编码*** ：将类别变量转换为数值形式，常见方法包括 `One-Hot`编码(独热编码)和标签编码

7. ***文本处理*** ：对文本数据进行清洗，如取出停用词、词干化、分词等

8. ***数据抽样*** ：从数据集中抽取样本，或通过采样/欠采样处理类别不平衡

9. ***特征工程*** ：创建新特征、删除不相关特征、选择重要特征等

### 1. Pandas 清洗空值

1. `pd.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)`

参数说明：

`axis`：默认为 0，表示逢空值剔除整行，如果设置参数 axis＝1 表示逢空值去掉整列。  
`how`：默认为 'any' 如果一行（或一列）里任何一个数据有出现 NA 就去掉整行，如果设置 how='all' 一行（或列）都是 NA 才去掉这整行。  
`thresh`：设置需要多少非空值的数据才可以保留下来的。  
`subset`：设置想要检查的列。如果是多个列，可以使用列名的 list 作为参数。  
`inplace`：如果设置 True，将计算得到的值直接覆盖之前的值并返回 None，修改的是源数据。

2. `df.isnull()` : 判断各个单元格是否为空
3. `df.fillna({'col' :value(mean,median,mode)}, inplace=True)` : 替换空字段

### Pandas 清洗重复数据

`df.duplicated()` 和 `df.drop_duplicates()` 方法

`df.duplicated()` 返回布尔值(第一个值返回True，之后的重复值False)

````python
person = {
  "name": ['Google', 'Runoob', 'Runoob', 'Taobao'],
  "age": [50, 40, 40, 23]  
}
df = pd.DataFrame(person)
display(df)
display(df.duplicated())
df.drop_duplicates(inplace=True)
display(df)
````

### Pandas 清洗错误数据

`df.loc[index, 'col'] = value` (单个值处理)

`df.drop(index, inplace=True)` (删除含错误值的行)

`df.replace(old_value, new_value)` : 将指定值替换为新值

````python
person = {
  "name": ['Google', 'Runoob' , 'Taobao'],
  "age": [50, 40, 12345]    # 12345 年龄数据是错误的
}

df = pd.DataFrame(person)
display(df)
# 错误值（单个）处理
df.loc[2, 'age'] = 30
display(df)

# 删除含错误值的行
for x in df.index:
    if df.loc[x, 'age'] > 120:
        df.drop(x, inplace=True)
display(df)

# 极端值处理
for x in df.index:
    if df.loc[x, 'age'] > 120:
        df.loc[x, 'age'] = 120
display(df)
````

以下是你提到的这些数据预处理和特征工程相关方法的详细讲解：

### 1. Z-score（标准化）(scipy.stats.zscore()或sklearn.preprocessing)
- **原理**：将数据转换为均值为0、标准差为1的分布，公式为 `z = (x - μ) / σ`，其中μ是均值，σ是标准差
- **用途**：消除不同特征的量纲影响，使数据处于同一量级，适用于近似正态分布的数据
- **适用场景**：SVM、线性回归等对特征尺度敏感的算法

### 2. IQR（四分位距）(pandas.Series.quantile()用于计算四分位数)
- **原理**：第75百分位数与第25百分位数的差值（IQR = Q3 - Q1）
- **用途**：主要用于异常值检测，通常将小于Q1-1.5×IQR或大于Q3+1.5×IQR的值视为异常值
- **优点**：对异常值不敏感，比均值和标准差更稳健

### 3. StandardScaler()(sklearn.preprocessing)
- **原理**：sklearn中的标准化工具，本质上实现Z-score标准化
- **特点**：
  - 计算训练集的均值和标准差用于缩放数据
  - 保持数据的分布形状，使各特征具有相同的方差
- **使用示例**：
```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

### 4. MinMaxScaler()(sklearn.preprocessing)
- **原理**：将数据缩放到[0,1]区间，公式为 `X_scaled = (X - X.min) / (X.max - X.min)`
- **特点**：
  - 会受到异常值的影响
  - 适用于分布较为均匀的数据
- **用途**：当需要将特征压缩到特定范围时使用，如神经网络的输入层

### 5. LabelEncoder()(sklearn.preprocessing)
- **原理**：将类别特征转换为0到n_classes-1的整数
- **适用场景**：处理有序类别特征（如学历：高中=0，本科=1，硕士=2）
- **注意**：不适合无序类别特征，会引入不必要的顺序关系
- **使用示例**：
```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(categorical_data)
```

### 6. pd.get_dummies()
- **原理**：对类别特征进行独热编码，为每个类别创建一个二进制列
- **适用场景**：处理无序类别特征（如颜色：红、绿、蓝）
- **特点**：
  - 不会引入顺序关系
  - 可能导致维度灾难（高基数特征）
- **使用示例**：
```python
import pandas as pd
encoded_data = pd.get_dummies(data, columns=['categorical_column'])
```

### 7. nltk.stem.PorterStemmer()(nltk.tem)
- **原理**：Porter词干提取算法，将单词还原为其词干形式
- **用途**：自然语言处理中减少词汇量，将不同形式的同一词归一化
- **示例**："running" → "run"，"flies" → "fli"
- **使用示例**：
```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_word = stemmer.stem("running")  # 结果为 "run"
```

### 8. nltk.word_tokenize()(nltk.tokenize)
- **原理**：将文本分割为单词或子词（tokens）的过程
- **用途**：NLP预处理的基础步骤，将连续文本转换为离散单元以便后续处理
- **示例**："I love NLP!" → ["I", "love", "NLP", "!"]
- **使用示例**：
```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize("Hello world!")  # 结果为 ["Hello", "world", "!"]
```

### 9. SMOTE()（上采样）(imblearn.over_sampling)
- **原理**：合成少数类样本的过采样技术，通过在少数类样本之间插值生成新样本
- **用途**：解决类别不平衡问题
- **优点**：不会导致过拟合（相比简单复制样本）
- **使用示例**：
```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

### 10. RandomUnderSampler()（下采样）(imblearn.under_sampling)
- **原理**：随机删除多数类样本，使各类别样本数量接近
- **用途**：解决类别不平衡问题
- **注意**：可能会丢失多数类中的重要信息
- **适用场景**：多数类样本量极大的情况

### 11. SelectKBest()(sklearn.feature_selection)
- **原理**：根据统计测试选择k个最佳特征
- **常用评分函数**：
  - 分类问题：卡方检验、互信息
  - 回归问题：皮尔逊相关系数、互信息
- **用途**：特征选择，减少维度，提高模型效率
- **使用示例**：
```python
from sklearn.feature_selection import SelectKBest, chi2
selector = SelectKBest(chi2, k=10)
X_new = selector.fit_transform(X, y)
```

### 12. PolynomialFeatures()(sklearn.preprocessing)
- **原理**：生成特征的多项式组合和交互项
- **参数**：degree指定多项式阶数
- **示例**：degree=2时，特征[x1, x2]会生成[x1, x2, x1², x2², x1x2]
- **用途**：捕捉特征之间的非线性关系和交互效应
- **注意**：可能导致维度爆炸，通常配合正则化使用

这些方法在机器学习 pipeline 中经常组合使用，根据具体数据特点和任务需求选择合适的预处理方法是构建高性能模型的关键步骤。

````python
# 独热编码(相当于向量表示类别)
data = {'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']}

df = pd.DataFrame(data)
display(df)
# 对 “City” 列进行 One-Hot 编码
df_encoded = pd.get_dummies(df, columns=['City'])
display(df_encoded)
````

````python
# 标准化
from sklearn.preprocessing import StandardScaler

data = {'Age': [25, 30, 35, 40, 45],
        'Salary': [50000, 60000, 70000, 80000, 90000]}

df = pd.DataFrame(data)
display(df)
# 标准化数据
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)  # 返回ndarray
print(df_scaled)
````
