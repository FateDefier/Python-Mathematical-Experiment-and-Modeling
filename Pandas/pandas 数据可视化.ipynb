# Pandas 数据可视化

Pandas 提供了与 Matplotlib 和 Seaborn 等可视化库的集成，使得数据的可视化变得简单而高效

| 图表类型 | 描述 | 方法 |
| --- | --- | --- |
| 折线图 | 展示数据随时间或其他连续变量的变化趋势 | `df.plot(kind='line')` |
| 柱状图 | 比较不同类别的数据 | `df.plot(kind='bar')` |
| 水平柱状图 | 比较不同类别的数据，但柱子水平排列 | `df.plot(kind='barh')` |
| 直方图 | 显示数据的分布 | `df.plot(kind='hist')` |
| 散点图 | 展示两个数值型变量之间的关系 | `df.plot(kind='scatter', x='col1', y='col2')` |
| 箱线图 | 显示数据分布，包括中位数、四分位数等 | `df.plot(kind='box')` |
| 密度图 | 展示数据的密度分布 | `df.plot(kind='kde')` |
| 饼图 | 显示不同部分在整体中的占比 | `df.plot(kind='pie')` |
| 区域图 | 展示数据的累计数值 | `df.plot(kind='area')` |

Pandas 数据可视化的基本功能和方法可以满足大多数日常数据可视化的需求，但若要实现更复杂的可视化，可以结合 Matplotlib 和 Seaborn 使用，进行更精细的图表定制。

调用格式: 

`df.plot(kind='figure_type', x, y, title, xlabel, ylabel, color, figsize, legend)`

### 1. 折线图(Line Plot)

通常用于展示 数据随时间的变化趋势

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Year': [2015, 2016, 2017, 2018, 2019, 2020],
        'Sales': [100, 150, 200, 170, 300, 250]}
df = pd.DataFrame(data)

df.plot(kind='line', x='Year', y='Sales', title='Sales vs Year', xlabel='Year', ylabel='Sales', figsize=(10, 8))
plt.show()
```

### 2. 柱状图(Bar Chart)

用于展示 不同分类间的比较，尤其适用于离散数据


```python
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [10, 15, 7, 12]}
df = pd.DataFrame(data)

df.plot(kind='bar', x='Category', y='Value', title='Category vs Value', xlabel='Category', ylabel='Value', figsize=(8, 5))
plt.show()
```

### 3. 散点图(Scatter Plot)

用于展示 两个数值变量间的关系

```python
data = {'Height': [150, 160, 170, 180, 190],
        'Weight': [50, 60, 70, 80, 90]}
df = pd.DataFrame(data)

df.plot(kind='scatter', x='Height', y='Weight', title='Height vs Weight', xlabel='Height(cm)', ylabel='Weight(kg)', figsize=(8, 5))
plt.show()
```

### 4. 直方图(Histogram)

用于展示 数据的分布，特别是数据的频率

```python
data = {'Scores': [55, 70, 85, 90, 60, 75, 80, 95, 100, 65]}
df = pd.DataFrame(data)

df.plot(kind='hist', y='Scores', bins=5, alpha=0.7, title='Scores Distribution', xlabel='Scores', ylabel='Frequency', edgecolor='k', figsize=(8, 5))  # y轴为频数
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# 手动设置分箱范围
custom_bins = [50, 60, 70, 80, 90, 100]
plt.figure(figsize=(8, 5))
df['Scores'].plot(kind='hist', bins=custom_bins, alpha=0.7, color='lightgreen', edgecolor='black')
plt.title('Scores Distribution with Custom Bins')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### 5. 箱线图(Box Plot)

用于展示 数据的分布情况，包括中位数、四分位数以及异常值

```python
data = {'Scores': [55, 70, 85, 90, 60, 75, 80, 95, 100, 65, 150]}
df = pd.DataFrame(data)

df.plot(kind='box', title='Scores Boxplot', ylabel='Scores', figsize=(8, 5))
plt.show()
```

### 6. 饼图(Pie Chart)

用于展示 各部分占总体的比例


```python
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [10, 15, 7, 12]}
df = pd.DataFrame(data)

df.plot(kind='pie', y='Value', labels=df['Category'], autopct='%1.1f%%', title='Category Proportions', figsize=(8, 5))
plt.show()
```

# Seaborn 与 Pandas 配合

### 7. 热力图

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

# 数据集中所有数值特征之间的散点图矩阵

`sns.pairplot()`

```python
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

sns.pairplot(df)
plt.show()
```
