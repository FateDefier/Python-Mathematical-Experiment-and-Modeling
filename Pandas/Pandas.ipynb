# 行列过滤

语法 dataframe[dataframe['col'] + 条件]

下面是单个条件，注意不要漏了dataframe和中括号

注意：条件在df[]的括号外面，即df[''] + 条件

```python
df[df['age'] > 70]
```

下面是多个条件

dataframe[(dataframe['col'] + 条件) 逻辑运算符 (dataframe['col' + 条件])]  

位运算符（|（按位或，多个条件中只要满足其中一个）&（按位与，多个条件同时满足））连接的两个条件用小括号括起来，外面用中括号

按位与 多个条件中只要满足其中一个

```python
data[(data['age'] > 20) | (data['age'] < 40)]  # 714行
```

按位或，多个条件同时满足

```python
data[(data['age'] > 20) & (data['age'] < 40)]  # 372行
```

筛选行和列

df[(df['col']+条件)][['col', 'col']]

行筛选一个条件也要加小括号，列筛选 筛选多列时两个中括号，筛选一列时一个中括号，当然，也可以先筛选行，后筛选列，通常只筛选行

```python
data[(data['age'] > 70)]['age']
```

```python
data[(data['age'] > 70)][['age', 'survived']]
```

筛选列

```python
data['age']
```

只查看一列的前五行(实际一个中括号展示应该没有列名)，返回的是Series(5,)

```python
data['age'].head()
```

只查看一列的前五行(实际两个中括号展示有列名)，返回的是dataframe(5,1)（数据框）

```python
data[['age']].head()
```

同样的，也可以筛选两列

此时，要使用两个中括号

```python
data[['age', 'survived']]
```

同时筛选行和列，建议使用 loc （location）方法

df.loc[(df['col']+条件 位运算符 (df['clo']+条件)), ['col', 'col']]

筛选多个参数是加中括号（返回DataFrame类型），1个参数时默认返回Series类型，要返回DataFrame类型加个中括号

```python
data.loc[(data['age'] > 70) & (data['survived'] > 0), ['survived', 'age']]
```

# pandas 重要应用：表格关联

result = pd.merge(
    
    left, 
    
    right, 
    
    how='inner', 
    
    on=None, 
    
    left_on=None, 
    
    right_on=None,
                  
    left_index=False, 
        
    right_index=False, 
        
    sort=False,
                  
    suffixes=('_x', '_y'), 
       
    copy=True, 
    
    indicator=False,
                  
    validate=None
)

参数说明
1. `left` 和 `right` ：必选参数，分别是要合并的两个 DataFrame 对象

2. `how`：指定合并的方式，默认为 'inner'。可选值有：

`inner`：内连接，只保留两个表中键值都存在的行，类似于 SQL 中的 INNER JOIN  
`outer`：外连接，保留两个表中的所有行，缺失值用 NaN 填充，类似于 SQL 中的 FULL OUTER JOIN  
`left`：左连接，保留左表中的所有行，右表中不存在匹配键值的行用 NaN 填充，类似于 SQL 中的 LEFT JOIN  
`right`：右连接，保留右表中的所有行，左表中不存在匹配键值的行用 NaN 填充，类似于 SQL 中的 RIGHT JOIN

3. `on`：指定用于合并的列名，该列必须在两个 DataFrame 中同时存在。如果两个 DataFrame 中列名不同，不能使用该参数，而应使用 left_on 和 right_on

4. `left_on` 和 `right_on`：分别指定左表和右表中用于合并的列名。当两个 DataFrame 中列名不同但表示相同含义时使用

5. `left_index` 和 `right_index`：如果为 True，则使用左表或右表的索引作为连接键。可以同时设置为 True 进行索引合并

6. `sort`：默认为 False，表示是否按照字典顺序对合并后的 DataFrame 进行排序

7. `suffixes`：当两个 DataFrame 中有相同列名时，用于区分的后缀，默认为 ('_x', '_y')

8. `copy`：默认为 True，表示是否复制数据

9. `indicator`：如果设置为 True 或字符串，会在结果中添加一列，指示每行数据的来源

10. `validate`：用于指定合并的约束条件，如 'one_to_one'（一对一连接）、'one_to_many'（一对多连接）等

rename()方法

DataFrame.rename(mapper=None, index=None, columns=None, axis=None, copy=True, inplace=False)


常用参数说明

mapper：用于指定映射关系，可以是字典或函数，axis和index或columns参数二选一使用，若同时指定，优先使用index或columns

当axis=0（即针对行索引操作）时，mapper是行索引的映射；

当axis=1（即针对列操作）时，mapper是列名的射

index：专门用于重命名行索引，值为字典或函数，字典形式为{原索引名: 新索引名}

columns：专门用于重命名列，值为字典或函数，字典形式为{原列名: 新列名}

axis：指定重命名操作的轴，取值为0（或'index'）表示对行索引操作，取值为1（或'columns'）表示对列进行操作

copy：默认为True，表示是否复制数据。如果设置为False，在某些情况下可能会返回一个视图而不是副本

inplace：默认为False，表示是否在原DataFrame上进行修改

若设置为True，则直接修改原DataFrame，方法返回None；

若设置为False，则返回一个修改后的新DataFrame，原DataFrame保持不变

```python
# survivel_x 和 survivel_y 不好看，修改列名
data_merge.rename(
    columns = {'survived_x': 'survived', 'survived_y': 'total'}, 
    inplace = True  # 在原数据框上修改
)
data_merge
```
