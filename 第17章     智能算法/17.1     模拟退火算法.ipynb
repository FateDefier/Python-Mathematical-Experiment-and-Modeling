例 17.1 已知 100 个目标的经度、纬度如表 17.1 所示. 我方有一个基地, 经度和纬度为 $(70, 40)$. 假设我方飞机的速度为 $1000\mathrm{km/h}$. 我方派一架飞机从基地出发, 侦察完所有目标, 再返回原来的基地. 在每一目标点的侦察时间不计, 求该架飞机所花费的时间 (假设我方飞机巡航时间可以充分长).  

这是一个**旅行商问题**. 给我方基地编号为 $0$, 目标依次编号为 $1, 2, \cdots, 100$, 最后我方基地再重复编号为 $101$ (这样便于在程序中计算). 距离矩阵 $D = (d_{ij})_{102 \times 102}$, 其中 $d_{ij}$ 表示 $i, j$ 两点的距离, $i, j = 0, 1, \cdots, 101$, 这里 $D$ 为实对称矩阵. 则问题是求一个从点 $0$ 出发, 走遍所有中间点, 到达点 $101$ 的一个最短路径.  

上面问题中给定的是大地坐标 (经度和纬度), 必须求两点间的实际距离. 设 $A, B$ 两点的大地坐标分别为 $(x_1, y_1), (x_2, y_2)$ (其中 $x_1, x_2$ 为经度, $y_1, y_2$ 为纬度), 过 $A, B$ 两点的大圆的劣弧长即为两点的实际距离. 以地心为坐标原点 $O$, 以赤道平面为 $XOY$ 平面, 以零度经线圈所在的平面为 $XOZ$ 平面建立三维直角坐标系. 则 $A, B$ 两点的直角坐标分别为  
$$
\begin{align*}
A & (R\cos x_1 \cos y_1, R\sin x_1 \cos y_1, R\sin y_1), \\
B & (R\cos x_2 \cos y_2, R\sin x_2 \cos y_2, R\sin y_2),
\end{align*}
$$  
其中 $R = 6370\mathrm{km}$ 为地球半径.

### 3. 模型及算法  
#### 1) 解空间  
解空间 $S$ 可表为 $\{0, 1, \cdots, 101\}$ 的所有固定起点和终点的循环排列集合, 即  
$$
\begin{align*}
S =& \left\{ (\pi_0, \pi_1, \cdots, \pi_{101}) \vert \pi_0 = 1, (\pi_1, \pi_2, \cdots, \pi_{100}) \text{ 为 } \{1, 2, \cdots, 100\} \right.\left. \text{的循环排列}, \pi_{101} = 101 \right\},
\end{align*}
$$  
其中每一个循环排列表示侦察 $100$ 个目标的一个回路, $\pi_i = j$ 表示在第 $i$ 次侦察目标 $j$, 初始解可选为 $(0, 1, \cdots, 101)$, 本文中先使用 Monte Carlo(蒙特卡罗) 方法求得一个较好的初始解.  


#### 2) 目标函数  
目标函数 (或称代价函数) 为侦察所有目标的路径长度. 要求  
$$
\min f(\pi_0, \pi_1, \cdots, \pi_{101}) = \sum_{i = 0}^{100} d_{\pi_i \pi_{i + 1}},
$$  
而一次迭代由下列三步构成.  


#### 3) 新解的产生  
设上一步迭代的解为 $\pi_0 \cdots \pi_{u - 1} \pi_u \pi_{u + 1} \cdots \pi_{v - 1} \pi_v \pi_{v + 1} \cdots \pi_{w - 1} \pi_w \pi_{w + 1} \cdots \pi_{101}$.  

(1) 2 变换法.  
任选序号 $u, v$, 交换 $u$ 与 $v$ 之间的顺序, 变成逆序($\pi_u \pi_{u + 1} \cdots \pi_{v - 1} \pi_v$逆序), 此时的新路径为  
$$
\pi_0 \cdots \pi_{u - 1} \pi_v \pi_{v - 1} \cdots \pi_{u + 1} \pi_u \pi_{v + 1} \cdots \pi_{101}.
$$  

(2) 3 变换法.  
任选序号 $u, v$ 和 $w$, 将 $u$ 和 $v$ 之间的路径插到 $w$ 之后(把$\pi_u \pi_{u + 1} \cdots \pi_{v - 1} \pi_v$插入到$\pi_w$后面), 对应的新路径为  
$$
\pi_0 \cdots \pi_{u - 1} \pi_{v + 1} \cdots \pi_w \pi_u \cdots \pi_v \pi_{w + 1} \cdots \pi_{101}.
$$  


#### 4) 代价函数差  
对于 2 变换法, 路径差可表示为  
$$
\Delta f = (d_{\pi_{u - 1} \pi_v} + d_{\pi_u \pi_{v + 1}}) - (d_{\pi_{u - 1} \pi_u} + d_{\pi_v \pi_{v + 1}}).
$$  


#### 5) 接受准则  
$$
P = 
\begin{cases} 
1, & \Delta f < 0, \\
\exp(-\Delta f / T), & \Delta f \geqslant 0,
\end{cases}
$$  
如果 $\Delta f < 0$, 则接受新的路径. 否则, 以概率 $\exp(-\Delta f / T)$ 接受新的路径, 即用计算机产生一个 $[0, 1]$ 区间上均匀分布的随机数 $\text{rand}$, 若 $\text{rand} \leqslant \exp(-\Delta f / T)$ 则接受.  


#### 6) 降温  
利用选定的降温系数 $\alpha$ 进行降温, 取新的温度 $T$ 为 $\alpha T$ (这里 $T$ 为上一步迭代的温度), 这里选定 $\alpha = 0.999$.  


#### 7) 结束条件  
用选定的终止温度 $e = 10^{-30}$, 判断退火过程是否结束. 若 $T < e$, 算法结束, 输出当前状态.  


### 4. 模型求解  
利用 Python 程序求得的巡航时间大约在 $43\mathrm{h}$ 左右, 其中的一个巡航路径如图 17.1 所示.  

代码结果改变的原因：

随机性本质：模拟退火算法是一种概率性算法，其核心在于通过随机扰动当前解来探索解空间。每次运行算法时，由于随机数的不同，得到的最终解可能会有所差异。

Metropolis准则：算法以一定概率接受较差的解，这有助于跳出局部最优解，但同时也意味着每次运行可能收敛到不同的解。

降温 schedule：你使用的降温系数（at = 0.999）和终止条件（e = 0.1**30）会导致算法在相对较少的迭代次数内终止（约13800次迭代后温度低于终止温度），这可能不足以让算法完全收敛到一个稳定的解。

初始解的影响：虽然你使用蒙特卡洛方法提供了一个较好的初始解，但模拟退火过程中的随机扰动仍然可能导致最终解的不同。

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.loadtxt("Pdata17_1.txt")
x = a[:, ::2].flatten()  # 经度, (100,)
y = a[:, 1::2].flatten()  # 纬度, (100,)
d1 = np.array([[70, 40]])  # (1, 2)二维数组
"""
np.c_默认(n,)是(n, 1)--column 列向量
np.r_默认(n,)是(1, n)--raw 行向量
"""
# 计算真实距离
xy = np.c_[x, y]  # (100, 2)
xy = np.r_[d1, xy, d1]  # 加入起点和终点0、1, (102, 2)
N = xy.shape[0]  # 102
t = np.radians(xy)  # 转化为弧度
for i in range(N):
    for j in range(N):
        d[i, j] = 6370 * np.lib.scimath.arccos(
            np.cos(t[i, 0] - t[j, 0]) * np.cos(t[i, 1]) * np.cos(t[j, 1]) + 
            np.sin(t[i, 1]) * np.sin(t[j, 1])  # cos(x1-x2)cosy1cosy2 + siny1siny2
        ).real  # 提取实部，(浮点数可能超出[-1, 1]的范围)所以np.lib.scimath.arccos可能得到复数
np.savetxt("实际距离数据.txt", np.c_[xy, d])  # 第一二列为x，y坐标，后面为102x102距离矩阵

# Monte Carlo得到一个较好的解
path = np.arange(N)  # 0~101地点标号
L = np.inf
for j in range(1000):  # 最大迭代次数
    path0 = np.arange(1, N-1)  # 1~100，观察点标号
    np.random.shuffle(path0)  # 打乱，直接改变原数组，只打乱第0维(块-->行-->列)
    path0 = np.r_[0, path0, N-1]  # (102,)
    L0 = d[0, path0[1]]  # 索引0号点到1号点的距离
    for i in range(1, N-1):
        L0 += d[path0[i], path0[i+1]]  # L0 + i号点到i+1号点的距离
    if L0 < L:
        path = path0
        L = L0
print(path, '\n', L)

# 模拟退火算法（包括Metropolis 采样算法）
e = 0.1 ** 30  # 终止温度
M = 20000  # 最大迭代次数
at = 0.999  # 新的温度的降温系数
T = 1  # 初始温度
for k in range(M):
    c = np.random.randint(1, 101, 2)  # 随机生成1~101的两个数，(2,)
    c.sort()
    c1 = c[0]  # 较小
    c2 = c[1]
    # path为蒙特卡洛得到的较好的路径
    # 2变换法 变换前[c1-1, c1] + [c2, c2+1]两个距离和逆序后[c1-1, c2] + [c1, c2+1]
    df=d[path[c1-1], path[c2]] + d[path[c1], path[c2+1]]-\
    d[path[c1-1], path[c1]] - d[path[c2], path[c2+1]]  # \Delta f(2变换法后的距离差)
    # Metropolis接受准则
    if df < 0:  # 接受
        path = np.r_[path[0], path[1:c1], path[c2:c1-1:-1], path[c2+1:102]]  # 注意，起点必须是0, path[c2:c1-1:-1]逆序
        L = L + df
    else:  # 概率选择
        if np.exp(-df / T) >= np.random.rand(1):  # 随机数小于概率时接受
            path = np.r_[path[0], path[1:c1], path[c2:c1-1:-1], path[c2+1:102]]
            L = L + df
    T = T * at  # 降温
    if T < e:  # 收敛准则
        break
print(path, '\n', L)  # 输出巡航路径及路径长度

xx = xy[path, 0]  # 102个点的经度
yy = xy[path, 1]  # 102个点的纬度

plt.rcParams['font.size'] = 16
plt.plot(xx, yy, '-*')
plt.show() 
```
