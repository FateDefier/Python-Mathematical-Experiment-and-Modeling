# 微分方程建模方法

#### 建立微分方程模型三步:
1. 根据实际要求确定研究的量（自变量、未知函数、必要的参数等），并确定坐标系
2. 找出这些量所满足的基本规律
3. 运用这些规律列出方程和定解条件

### 1. 按规律直接列方程

建立物体冷却过程的数学模型.

将某物体放置于空气中，在时刻$t = 0$测量得它的温度为$u_0 = 150^\circ\text{C}$，10min 后测量得它的温度为$u_1 = 100^\circ\text{C}$. 要求建立此物体的温度$u$和时间$t$的关系，并计算 20min 后物体的温度. 其中假设空气的温度保持为$\bar{u} = 24^\circ\text{C}$.

解 牛顿冷却定律是温度高于周围环境的物体向周围媒质传递热量逐渐冷却时所遵循的规律：当物体表面与周围存在温度差时，单位时间从单位面积散失的热量与温度差成正比，比例系数称为热传递系数.

假设该物体在时刻$t$的温度为$u = u(t)$，则由牛顿冷却定律，得到
$$
\frac{du}{dt} = -k (u - \bar{u}), \tag{8.4}
$$
其中，$k > 0$，方程 (8.4) 就是物体冷却过程的数学模型.

注意到$\bar{u} = 24$为常数，$u - \bar{u} > 0$，可将方程 (8.4) 改写为
$$
\frac{d(u - 24)}{u - 24} = -k dt.
$$

两边积分得到
$$
\int_{150}^{u} \frac{d(u - 24)}{u - 24} = \int_{0}^{t} -k dt,
$$
化简得
$$
u = 24 + 126 e^{-k t}. \tag{8.5}
$$

把条件$t = 10$, $u = u_1 = 100$代入式 (8.5)，得$k = \frac{1}{10} \ln \frac{126}{76} = 0.0506$，故此物体的温度$u$和时间$t$的关系为$u = 24 + 126 e^{-0.0506 t}$. 20 分钟后物体的温度为$69.8413^\circ\text{C}$.

```python
import sympy as sp

sp.var('t, k')  # 定义符号变量
u = sp.var('u', cls=sp.Function)  # 定义函数 u(t)
eq = sp.diff(u(t), t) + k * (u(t) - 24)  # 定义微分方程
uu = sp.dsolve(eq, ics={u(0):150})  # 求解微分方程，给定初值，uu是equality对象
print(uu.rhs)  # 打印解，.rhs（right hand side等式右边的解）

kk = sp.solve(uu, k)  # kk是一个列表，包含所有可能的 k 值，求解出 k 的表达式
k0 = kk[0].subs({t: 10, u(t): 100})  # 带入给定数据，计算 k 的特定值（即常数）
print(kk, '\t', k0, k0.n())  # 打印所有 k 值和特定 k 值（类似符号解和数值解）

# uu 是 Equality 对象；.args 是一个元组 (u(t), 24 + 126*exp(-k*t))
u1 = uu.args[1]  # 取第二个元素，即等式右边解的表达式
u0 = u1.subs({t:20, k: k0})  # 带入t和k0，计算 u(20) 的值
print(f"20min后的温度为：{u0:.2f}℃")  # 打印结果，保留两位小数
```

### 2. 微元分析法

该方法的基本思想是通过分析研究对象的有关变量在一个很短时间内的变化规律，寻找一些微元之间的关系式。

有高为 1m 的半球形容器，水从它的底部小孔流出，小孔横截面积为 1cm²，开始时容器内盛满了水，求水的高 h 随时间 t 变化的规律。

已知条件

容器为半球形，半径 r = 1m

水的初始高度 h = 1m

小孔的横截面积 S = 1cm² = 0.0001m²

重力加速度 g = 9.8m/s²

求解过程：

水的变化率，满足：

$$ Q = \frac{dV}{dt} = 0.62S \sqrt{2gh}, $$

式 (8.6) 中，0.62 为流量系数；g 为重力加速度 (取 9.8m/s²)，S 为孔口横截面积 (单位：m²)，h 为 t 时刻水面高度 (单位：cm)。当 S = 1cm² = 0.0001m² 时，

$$ dV = 0.000062 \sqrt{2gh}dt. $$

在微小时间间隔 [t, t+dt] 内，水面高度由 h 降到 h+dh(这里 dh < 0)，容器中水的体积变化量近似为

$$ dV = -\pi r^2 dh, $$

式 (8.8) 中，r 为 t 时刻的水面半径，右端量负号是由于 dh < 0，而 dV > 0；这里

$$ r^2 = [1^2 - (1 - h)^2] = 2h - h^2. $$

由式 (8.7)—(8.9)，得

$$ 0.000062 \sqrt{2gh}dt = \pi (h^2 - 2h) dh. $$

再考虑到初始条件，得到如下的微分方程模型

$$ \begin{cases} \frac{dt}{dh} = \frac{10000 \pi}{0.62 \sqrt{2g}}({h^{\frac{3}{2}} - 2h^{\frac{1}{2}}}), \ t(1) = 0, \end{cases} $$

利用分离变量法，可以求得微分方程的解为

$$ t(h) = -15260.5042h^{\frac{3}{2}} + 4578.1513h^{\frac{1}{2}} + 10682.3530. $$

上式表达了水从小孔流出的过程中容器内水面高度 h 与时间 t 之间的关系。

```python
import sympy as sp

h = sp.symbols('h')  # 定义符号变量 h
t = sp.symbols('t', cls=sp.Function)  # 定义函数 t(h)
g = 9.8  # 重力加速度
eq = t(h).diff(h) - 10000*sp.pi*(h**(3/2) - 2*h**(1/2))/(0.62 * sp.sqrt(2*g))  # 定义方程
t = sp.dsolve(eq, ics={t(1): 0})  # 求解方程符号解，给定初值
print(t.args[1].n(9))  # 打印解的数值，保留9位小数
```

### 3. 模拟近似法

该方法的基本思想是在不同的假设下模拟实际的现象，即建立模拟近似的微分方程， 从数学上求解或分析解的性质，再去和实际情况作对比，观察这个模型能否模拟、近似某些实际的现象．

例：（交通管理问题）在交通十字路口都会设置红绿灯．为了让那些正行驶在交叉路口或离交叉路口太近而无法停下的车辆通过路口，红绿灯转换中间还要亮起一段时间的黄灯那么， 黄灯应亮多长时间才最为合理呢？

分析: 黄灯状态持续的时间包含驾驶员的反应时间、车通过交叉路口的时间以及通过刹车距离所需的时间.

解 记 $v_0$ 是法定速度, $I$ 是交通路口的长度, $L$ 是典型的车身长度, 则车通过路口的时间为 $\frac{I + L}{v_0}$.

下面计算刹车距离, 刹车距离就是从开始刹车到速度 $v = 0$ 时汽车驶过的距离. 设 $W$ 为汽车的重量, $\mu$ 为摩擦系数. 显然, 地面对汽车的摩擦力为 $\mu W$, 其方向与运动方向相反. 由牛顿第二定律, 汽车在停车过程中, 行驶的距离 $x$ 与时间 $t$ 的关系可由下面的微分方程表示
$$
\frac{W}{g} \cdot \frac{d^2 x}{dt^2} = -\mu W,
$$
其中, $g$ 为重力加速度. 化简上式, 得
$$
\frac{d^2 x}{dt^2} = -\mu g.
$$

再考虑初始条件, 建立如下的二阶微分方程模型
$$
\begin{cases}
\frac{d^2 x}{dt^2} = -\mu g, \\
\left. x \right|_{t = 0} = 0,\quad \left. \frac{dx}{dt} \right|_{t = 0} = v_0.
\end{cases}
$$
先求解二阶微分方程 $\frac{d^2 x}{dt^2} = -\mu g$ 式, 对其两边从 $0$ 到 $t$ 积分, 利用初始条件得
$$
\frac{dx}{dt} = -\mu g t + v_0.
$$
再积分上式一次, 求得二阶微分方程初值问题的解为
$$
x(t) = -\frac{1}{2} \mu g t^2 + v_0 t.
$$
在 $\frac{dx}{dt} = -\mu g t + v_0$ 式中令 $\frac{dx}{dt} = 0$, 可得刹车所用时间 $t_0 = \frac{v_0}{\mu g}$, 从而得到刹车距离 $x(t_0) = \frac{v_0^2}{2 \mu g}$.

下面计算黄灯状态的时间 $T$, 则
$$
T = \frac{x(t_0) + I + L}{v_0} + T_0,
$$
其中 $T_0$ 是驾驶员的反应时间, 代入 $x(t_0)$ 得
$$
T = \frac{v_0}{2 \mu g} + \frac{I + L}{v_0} + T_0.
$$

设 $T_0 = 1 \mathrm{s}$, $L = 4.5 \mathrm{m}$, $I = 9 \mathrm{m}$. 另外, 取具有代表性的 $\mu = 0.7$, 当 $v_0 = 45 \mathrm{km/h}$, $65 \mathrm{km/h}$ 以及 $80 \mathrm{km/h}$ 时, 黄灯时间 $T$ 如表 8.1 所列.

```python
import numpy as np

v0 = np.array([45, 65, 80])  # 法定速度
T0 = 1  # 驾驶员反应时间
L = 4.5  # 典型车身长度
I = 9  # 交通路口长度
miu = 0.7  # 摩擦系数
g = 9.8  # 重力加速度
T = v0 / (2*miu*g) + (I + L) / v0 + T0  # 黄灯时间
print(T)  # 打印黄灯时间数组
```
