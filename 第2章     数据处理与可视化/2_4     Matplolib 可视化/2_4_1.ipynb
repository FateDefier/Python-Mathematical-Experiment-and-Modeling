```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 设置中文字体和负号显示
plt.rcParams.update({
    'font.family': ['SimHei'],
    'axes.unicode_minus': False,
    'font.size': 6  # 统一设置字体大小
})
```

```python
# 读取 Excel 文件并选择列（推荐使用列名而非索引）
df = pd.read_excel("excel1.xlsx", usecols=["用户A", "用户B", "用户C"])

# 计算每列的总和
total_values = df.sum()

# 创建柱状图
fig, ax = plt.subplots(figsize=(10, 10))  # 明确创建 Figure 和 Axes 对象
bars = ax.bar(
    x=total_values.index,  # 使用列名作为 x 轴标签
    height=total_values,   # 使用总和值作为柱子高度
    width=0.4,             # 调整柱子宽度
    color=['#3274A1', '#E1812C', '#3A923A'],  # 自定义颜色
    edgecolor='black',     # 添加边框
    alpha=0.8              # 设置透明度
)

# 设置图表标题和坐标轴标签
ax.set_title("用户消费数据对比", pad=15)
ax.set_ylabel("消费数据", labelpad=10)
ax.set_xlabel("用户", labelpad=10)

# 为每个柱子添加数值标签
ax.bar_label(bars, padding=3, fmt='%.1f')

# 设置 x 轴标签旋转角度（避免重叠）
plt.xticks(ha='right')  # ha参数设置水平对齐方式

# 设置 y 轴范围（可选）
ax.set_ylim(0, total_values.max() * 1.1)  # 稍微扩展 y 轴上限

# 添加网格线（可选）
ax.grid(axis='y', linestyle='--', alpha=0.7)

# 调整布局，确保标签完整显示
plt.tight_layout()

# 保存高分辨率图片
plt.savefig('figure_excel1.png', dpi=500, bbox_inches='tight')

# 显示图表
plt.show()
```
