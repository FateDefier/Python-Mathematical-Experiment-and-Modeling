```python
a = pd.read_excel("excel1.xlsx", usecols=range(1, 4))
b = a.values  # 提取其中的数据
c = a.describe()  # 对数据进行统计描述，返回数量，均值，标准差，最大最小值，25%、50%、75%分位数
print(b)
c  # 只能这样输出一行
```

```python
# 生成 DataFrame 类型数据
d = pd.DataFrame(b, index=np.arange(1, 11),
                 columns=["用户A", "用户B", "用户C"])
f = pd.ExcelWriter('excel2.xlsx')  # 创建文件对象
d.to_excel(f, "sheet1")  # 把 d 写入Excel文件中
d.to_excel(f, "sheet2")  # 把 d 再写入另一个表单中
f.close()  # 高版本 pandas 没有 save() 方法
```

```python
# 3.数据子集的获取
a = pd.read_excel("excel1.xlsx", usecols=range(1, 4))
b1 = a.iloc[np.arange(6), [0, 1]]  # 通过标号筛选数据
b2 = a.loc[np.arange(6), ["用户A", "用户B"]]  # 通过标签筛选数据
print(b1, '\n', b2)
```
