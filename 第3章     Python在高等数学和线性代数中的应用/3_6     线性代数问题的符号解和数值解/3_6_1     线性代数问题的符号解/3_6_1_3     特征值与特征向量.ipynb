# 特征值与特征向量

### sympy.Matrix.eigenvals()

| 参数名                     | 类型                 | 默认值     | 说明                                   |
| ----------------------- | ------------------ | ------- | ------------------------------------ |
| `error_when_incomplete` | `bool`             | `True`  | 若无法求出全部特征值，是否抛出 `MatrixError`。       |
| `simplify`              | `bool \| Callable` | `False` | 是否对特征多项式或根进行化简；可为布尔值或自定义化简函数。        |
| `multiple`              | `bool`             | `False` | 为 `True` 时返回列表（含重根），为 `False` 时返回字典。 |
| `rational`              | `bool`             | `True`  | 是否尝试先用有理算法求特征多项式，再求根。                |

| 返回值         | 类型                                                     | 说明                        |
| ----------- | ------------------------------------------------------ | ------------------------- |
| eigenvalues | `dict[Expr, int]`（默认）<br>`list[Expr]`（`multiple=True`） | 特征值到代数重数的映射，或按重数展开的特征值列表。 |


### sympy.Matrix.eigenvects()

| 参数名                     | 类型                       | 默认值                   | 说明                              |
| ----------------------- | ------------------------ | --------------------- | ------------------------------- |
| `error_when_incomplete` | `bool`                   | `True`                | 若无法求出全部特征向量，是否抛出 `MatrixError`。 |
| `iszerofunc`            | `Callable[[Expr], bool]` | `lambda x: x.is_zero` | 判断一个表达式是否为零的回调函数，用于消元。          |
| `simplify`              | `bool \| Callable`       | `True`                | 是否对特征向量进行化简；可为布尔值或自定义化简函数。      |
| `chop`                  | `bool \| float`          | `False`               | 对浮点结果做截断误差处理；可设为阈值或布尔值。         |

| 返回值        | 类型                                     | 说明                                                            |
| ---------- | -------------------------------------- | ------------------------------------------------------------- |
| eigenvects | `list[tuple[Expr, int, list[Matrix]]]` | 每个元素为 `(特征值, 代数重数, [线性无关特征向量列表])` 的三元组。特征向量类型为 `Matrix`（列向量）。 |



求下列矩阵的特征值和特征向量

$$
A = \begin{bmatrix}
0 & -2 & 2 \\
-2 & -3 & 4 \\
2 & 4 & -3
\end{bmatrix}
$$

```python
import sympy as sp
A = sp.Matrix([[0, -2, 2], [-2, -3, 4], [2, 4, -3]])
print(f"A的特征值为：{A.eigenvals()}")
print(f"A的特征向量为：{A.eigenvects()}")
```

### 2. 相似对角化

### Matrix.diagonalize()

| 参数名          | 类型              | 默认值     | 说明                                                                                         |
| ------------ | --------------- | ------- | ------------------------------------------------------------------------------------------ |
| `normalvecs` | `bool`          | `False` | 如果为 `True`，返回的特征向量会被单位化（L2 范数为 1）。                                                         |
| `sort`       | `bool \| str`   | `True`  | 如何对特征值/特征向量排序：<br>`True` → 按实部从小到大；<br>`">"` / `"<"` → 显式指定大小顺序；<br>`False` → 不排序（保持内部顺序）。 |
| `chop`       | `bool \| float` | `False` | 浮点结果是否做截断误差处理（可给定阈值）。                                                                      |

| 返回值      | 类型                      | 说明                                              |
| -------- | ----------------------- |-------------------------------------------------|
| `(P, D)` | `tuple[Matrix, Matrix]` | `P` 的列是特征向量，P为特征矩阵`D` 是对角矩阵，满足 **A = P·D·P⁻¹**。 |
| 或抛出      | `MatrixError`           | 若矩阵不可对角化（几何重数 < 代数重数）则抛出异常。                     |


    把下列矩阵相似对角化，即求可逆矩阵$P$，使得$P^{-1}AP=D$为对角矩阵

$$
\boldsymbol{A} = \begin{bmatrix}
0 & -2 & 2 \\
-2 & -3 & 4 \\
2 & 4 & -3
\end{bmatrix}
$$

```python
from sympy import Matrix
A = Matrix([[0, -2, 2], [-2, -3, 4], [2 , 4, -3]])
if A.is_diagonalizable():
    print(f"A的对角化矩阵为：\n{A.diagonalize()}")
else:
    print(f"A不能对角化")
```
