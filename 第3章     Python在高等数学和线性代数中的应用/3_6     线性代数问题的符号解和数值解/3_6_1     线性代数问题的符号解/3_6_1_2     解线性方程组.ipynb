# 解线性方程组

### 1. 解唯一
求下列非线性方程组的符号解

$$
\begin{cases}
2x_1 + x_2 - 5x_3 + x_4 = 8, \\
x_1 - 3x_2 - 6x_4 = 6, \\
2x_2 - x_3 + 2x_4 = -2, \\
x_1 + 4x_2 - 7x_3 + 6x_4 = 2.
\end{cases}
$$

记上述线性方程组为 $Ax=b$，可以验证系数矩阵 $A$ 的秩$R(A) = 4$，所
以线性方程组有唯一解$x＝A^-1b$（两边同时左乘$A^-1$可求解x），利用 Python 软件求得
$$ x=[4, \frac{-14}{9}, \frac{-2}{9}, \frac{4}{9}]^T $$

```python
import sympy as sp
A = sp.Matrix([[2, 1, -5, 1], [1, -3, 0, -6], [0, 2, -1, 2], [1, 4, -7, 6]])
b = sp.Matrix([8, 6, -2, 2])  # 行向量
b.transpose()  # 建议用tranpose，使用.T会有警告，也可以直接输入列向量
print(f"系数矩阵的秩为：{A.rank()}")
print(f"线性方程组的解为：{A.inv()*b}")  # 左乘A逆
```

### 2. 基础解系（解无穷）（齐次通解）

| 参数名           | 类型                           | 说明                                                        |
| ------------- | ---------------------------- | --------------------------------------------------------- |
| *self*        | `MatrixBase`                 | 调用该方法的矩阵对象（通常为 `Matrix`、`ImmutableMatrix` 等）。             |
| `simplify`    | `bool`，可选                    | 计算过程中是否对中间结果做化简，默认为 `False`。设为 `True` 可能提高结果可读性，但会牺牲部分性能。 |
| `iszerofunc`  | `Callable[[Basic], bool]`，可选 | 判断某个表达式是否视为零的回调函数。默认使用 SymPy 内部的 `_iszero`。               |
| `maxrank`     | `int`，可选                     | 可提前给出的最大秩上限，用于提前终止计算；默认为 `None`（不限制）。                     |
| `normalize`   | `bool`，可选                    | 若设为 `True`，则把得到的零空间基向量单位化（L2 范数为 1）；默认为 `False`。          |
| `dotprodsimp` | `bool`，可选                    | 是否在 Gram-Schmidt 正交化过程中使用点积化简；默认为 `None`（由 SymPy 内部策略决定）。 |  

| 返回值         | 类型             | 说明                                                    |
| ----------- | -------------- | ----------------------------------------------------- |
| null\_space | `list[Matrix]` | 由 `Matrix` 对象组成的列表，每个元素都是零空间的一组基向量。若矩阵满秩，则返回空列表 `[]`。 |

求下列齐次线性方程组的基础解系

$$
\begin{cases}
x_1 - 5x_2 + 2x_3 - 3x_4 = 0\\
5x_1 + 3x_2 + 6x_3 - x_4 = 0\\
2x_1 + 4x_2 + 2x_3 + x_4 = 0
\end{cases}
$$

```python
import sympy as sp
A = sp.Matrix([[1, -5, 2, -3], [5, 3, 6, -1], [2, 4, 2, 1]])
print(f"A的零空间（即基础解系）为：'\n'{A.nullspace()}")
```

### 3.非齐次方程组的通解

首先，将非齐次方程转化成齐次方程，然后，使用 rref()把增广矩阵化成最简形，判断基础解系

求下列非齐次线性方程组的通解

$$
\begin{cases}
x_1 + x_2 - 3x_3 - x_4 = 1, \\
3x_1 - x_2 - 3x_3 + 4x_4 = 4, \\
x_1 + 5x_2 - 9x_3 - 8x_4 = 0.
\end{cases}
$$

```python
import sympy as sp
A = sp.Matrix([[1, 1, -3, -1], [3, -1, -3, 4], [1, 5, -9, -8]])
b = sp.Matrix([[1], [4], [0]])
C = A.row_join(b)
print(f"增广矩阵的最简形为: \n{C.rref()}")
# 使用nullspace()只能得到齐次的通解，并不能求出非齐次的特解
# print(C.nullspace())
# 后面可以根据最简形手动求解
```
