# 求超定线性方程组的最小二乘解

建模中经常使用线性最小二乘法，实际上就是求超定线性方程组（未知数个数少， 方程个数多）的最小二乘解， 前面已经使用 pinv(）求超定线性方程组的最小二乘解（矛盾方程），下面再举两个求最小二乘解的例子，并使用 numpy.linalg模块的 lstsq(）函数求解.

求超定线性方程组
$$
\begin{bmatrix}
0 & 1 \\
1 & 1 \\
2 & 1 \\
3 & 1
\end{bmatrix}
\begin{bmatrix}
m \\
c
\end{bmatrix}
=
\begin{bmatrix}
-1 \\
0.2 \\
0.9 \\
2.1
\end{bmatrix}
$$
的最小二乘解，相当于给定$x$的观测值$\textbf{x}=[0, 1, 2, 3]^T$，$y$的观测值$\textbf{y}=[-1, 0.2, 0.9, 2.1]^T$，拟合经验函数$y=mx+c$

```python
import numpy as np
import numpy.linalg as LA
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3])  # 一维数组。不是行向量
y = np.array([-1, 0.2, 0.9, 2.1])
# np.c_ 把--一维数组--视为 列向量
A = np.c_[x, np.ones_like(x)]
print(A)
m, c = LA.lstsq(A, y, rcond=None)[0]
print(m, c)
plt.rc('font', size=16)
plt.rc('font', family='SimHei')
plt.rc('axes', unicode_minus=False)
plt.plot(x, y, 'o', label='原始数据', markersize=5)
plt.plot(x, m*x+c, 'r', label='拟合直线')  # 注意这里是x，不是A，否则维度不一致
plt.legend()
plt.savefig("回归.png", dpi=500)
plt.show()
```

$$
\text{为了测量刀具的磨损速度，做这样的实验：经过一定时间（如每隔一小时），测量一次刀具的厚度，得到一组实验数据 } (t_i, y_i) \ (i = 1, 2, \cdots, 8) \ \text{如表 3.3 所示. 试用实验数据拟合经验公式 } y = at + b.
$$

$$
\textbf{表 3.3} \quad \text{实验数据}
$$

$$
\begin{array}{c|c|c|c|c|c|c|c|c}
\hline
t_i & 0 & 1 & 2 & 3 & 4 & 5 & 6 & 7 \\
\hline
y_i & 27.0 & 26.8 & 26.5 & 26.3 & 26.1 & 25.7 & 25.3 & 24.8 \\
\hline
\end{array}
$$

A.dot(ab),类似

$$
\begin{bmatrix}
0 & 1 \\
1 & 1 \\
2 & 1 \\
3 & 1
\end{bmatrix}
\begin{bmatrix}
m \\
c
\end{bmatrix}
=
\begin{bmatrix}
-1 \\
0.2 \\
0.9 \\
2.1
\end{bmatrix}
$$

A即$\begin{bmatrix}
0 & 1 \\
1 & 1 \\
2 & 1 \\
3 & 1
\end{bmatrix}$
ab即$\begin{bmatrix}
m \\
c
\end{bmatrix}$
做矩阵乘法求得y

```python
import numpy as np
import numpy.linalg as LA
import matplotlib.pyplot as plt
t = np.arange(8)
y = np.array([27.0, 26.8, 26.5, 26.3, 26.1, 25.7, 25.3, 24.8])
A = np.c_[t, np.ones_like(t)]
ab = LA.lstsq(A, y, rcond=None)[0]  # 返回值为向量
print(ab)
plt.rc('font', size=16)
plt.plot(t, y, 'o', label='Original data', markersize=5)
plt.plot(t, A.dot(ab), 'r', label='Fitted line')
plt.legend()
plt.savefig("刀具损耗.png", dpi=500)
plt.show()
```
