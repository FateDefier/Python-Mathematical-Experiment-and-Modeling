求极大值的函数  
SciPy 没有专门的“求极大值”函数，通用做法是：  
对目标函数取负值，再调用最小化器。  

| 场景       | 推荐方式                            | 关键代码示例                                                                    |
| -------- | ------------------------------- | ------------------------------------------------------------------------- |
| 单变量有界求极大 | `minimize_scalar` 或 `fminbound` | `res = minimize_scalar(lambda x: -f(x), bounds=(a, b), method='bounded')` |
| 多维无约束求极大 | `fmin` 或 `minimize`             | `res = fmin(lambda x: -f(x), x0)`                                         |


求函数 $f(x)=e^x\cos(2x)$ 在区间 $[0,3]$ 上的极小值点

```python
import numpy as np
from scipy.optimize import fminbound
f = lambda x: np.exp(x)*np.cos(2*x)
x0 = fminbound(f, 0, 3)
print(f"极小值点为：{x0}\n极小值为：{f(x0)}")
```

求函数 $f(x)=e^x\cos(2x)$ 在 0 附近的一个极小点

```python
from numpy import exp, cos
from scipy.optimize import fmin
f = lambda x: exp(x)*cos(2*x)
x0 = fmin(f, 0)
print(f"极小值点为：{x0}\n极小值为：{f(x0)}")
```

2. 多元函数的极值点

先简单介绍scipy.optimize 模块下的 minimize 函数，该函数求多元函数的极小
值点和极小值， 其基本调用格式为  
#### minimize (fun, x0, args=(), method=None)  
其中， method 表示使用的算法其取值可为  
'Nelder-Mead','Powell','CG'/'BFGS','Newton-CG','L-BFGS-B', 
'TNC','COBYLA','SLSQP','trust-constr','dogleg','trust-ncg','trust-exact' , 'trust-krylov'. 

返回数组，取x（极小值点）用x0.x,去函数值（极小值）用x0.fun

| 函数 | 官方签名（常用参数高亮） | 默认值 | 返回值说明 |
|---|---|---|---|
| minimize | `scipy.optimize.minimize(fun, x0, **args)` | — | OptimizeResult |
| | **fun** 目标函数 | — | |
| | **x0** 初始猜测 | — | |
| | **method** 算法 | 'BFGS' | |
| | **jac** 梯度函数 | None | |
| | **bounds** 变量界限 | None | |
| | OptimizeResult.x | — | ndarray 最优解 |

求函数 $ f(x)=100(x_2-x_1^2)^2+(1-x_1)^2 $ 的极小值

```python
from scipy.optimize import minimize
f = lambda x:100*(x[1]-x[0]**2)**2+(1-x[0])**2
x0 = minimize(f, [2.0, 2.0])
print(f"极小值点为：{x0.x}\n极小值为：{x0.fun}")
```
