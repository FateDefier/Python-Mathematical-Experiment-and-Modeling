# 数值积分 

## 1.一重积分
  在实际问题中， 利用牛顿－莱布尼兹公式，通过求原函数，计算定积分

$$ \int_{a}^{b} f(x) \, dx = \left. F(x) \right|_{a}^{b} = F(b) - F(a) $$

是非常困难的， 这里F(x) 为 f(x) 的一个原函数．然而，我们知道，当一元函数 f(x)在区间 [a,b] 上不变号时，其定积分的值恰好等于 f(x) 与直线 x = a, x = b 以及 x轴所围成的曲边梯形的“有向面积”·

  梯形法的计算思想是把大的曲边梯形剖分成多个小的曲边梯形， 每个小曲边梯形的面积用一个梯形面积作近似，最后累加求和得到所求定积分的数值解．

  下面直接给出积分 $ \int_{a}^{b} f(x) \, dx $ 的梯形计算公式和辛普森计算公式.

  使用梯形公式时， 把整个区间 [a,b] 进行n等分(此时步长$ h= (b-a)/n $)，分点为 $ x_i (i = 0,1, \cdots ,n) $ 对应的函数值 $ f_i = f(x_i) $， 这里 $ x_i=a+i·h , i=0,1, \cdots ,n $. 
积分$ \int_{a}^{b} f(x) \, dx $的数值解$$ I_1 = \frac{h}{2} \left( f_0 + 2 \sum_{i=1}^{n-1} f_i + f_n \right) $$

  使用辛普森公式（三个点一算1:4:1，2是重叠出来的）时将区间 [a, b] 进行 2n 等分，步长 $ h=\frac{b-a}{2n} $，分点 $ x_i=a+i·h(i = 0, 1, \cdots, 2n) $对应的函数值 $f_i = f(x_i)$．积分 $\int_{a}^{b} f(x)dx $ 的数值解
$$ I_2 = \frac{h}{3} \left[ f_0 + 2 \sum_{i=1}^{n-1} f_{2i} + 4 \sum_{i=1}^{n} f_{2i-1} + f_{2n} \right] $$

积分$$ \int_{0}^{1} \sin\left(\sqrt{\cos x + x^2}\right) dx $$
的数值解

## quad(function, a, b)

### quad 函数计算定积分，返回的第一个值为积分的结果，第二个值为误差

```python
import numpy as np
from scipy.integrate import quad
# 定义梯形公式的函数
def trapezoid(f0, n, a, b):  # trapezoid 不规则四边形
    xi = np.linspace(a, b, n)
    h = (b-a)/(n-1)
    return h*(np.sum(f0(xi))-(f0(a)+f0(b))/2)
# 定义辛普森公式的函数
# 奇数 odd number 偶数 even number
def simpson(f, n, a, b):
    xi = np.linspace(a, b, 2*n + 1)  # 分成 2n 段，包含端点，总共2n+1个点
    h = (b - a) / (2.0 * n)
    s = f(a) + f(b)
    j_sum = np.sum([f(xi[i]) for i in range(1, 2*n, 2)])  # 终止索引为2n，不包括 b 点
    o_sum = np.sum([f(xi[i]) for i in range(2, 2*n, 2)])  # 开始索引为 2，不包括 a 点
    s += 4*j_sum + 2*o_sum
    return s*h/3.0
# 等效下面
# def simpson(f, n, a, b):
#     xi = np.linspace(a, b, 2*n + 1)
#     h = (b - a) / (2.0 * n)
#     xe = [f(xi[i]) for i in range(len(xi)) if i%2 == 0]
#     xo = [f(xi[i]) for i in range(len(xi)) if i%2 != 0] 
#     return h / 3.0 * (2*np.sum(xe)+4*np.sum(xo)-f(a)-f(b))
f1 = lambda x: np.sin(np.sqrt(np.cos(x)+x**2))
print(f"梯形法积分的结果：{trapezoid(f1, 1000, 0, 1)}")
print(f"辛普森法积分的结果：{simpson(f1, 1000, 0, 1)}")
print(f"SciPy积分的结果：{quad(f1, 0, 1)}")
```

## 多重积分

关千多重积分使用 SciPy库中的函数dblquad, tplquad 直接求数值解 

### dblquad(func, a, b, gfun, hfun, args=(), epsabs=1.49e-08, epsrel=1.49e-08)  
    其中，被积函数 func 的格式为 func(y, x)，最外层 x 的积分区间为 [a, b]，内层 y 的积分区间为 [gfun(x), hfun(x)]

### tplquad(func, a, b, gfun, hfun, qfun, rfun, args=(), epsabs=1.49e-08, epsrel=1.49e-08)  
    其中，被积函数 func 的格式为 func(z, y, x)，最外层 x 的积分区间为 [a, b]，中间层y 的积分区间为 [gfun(x), hfun(x)］，最内层 z 的积分区间为 [qfun(x, y), rfun(x, y)].

(1)$\int_{0}^{2}dx \int_{0}^{1} xy^2dy$
(2)$\iint_{x^2+y^2 \leqslant 1} e^{-\frac{x^2}{2}}\sin(x^2+y)dx\,dy$

(2)中要先化成累次积分$$I = \int_{-1}^{1} dx\int_{\sqrt{1-x^2}}^{-\sqrt{1-x^2}} e^{-\frac{x^2}{2}} \sin(x^2+y)\,dy$$

```python
import numpy as np
from scipy.integrate import dblquad
f1 = lambda x, y: x*y**2
print(f"第一个二重积分 = {dblquad(f1, 0, 2, 0, 1)}")
f2 = lambda x, y: np.exp(-x**2/2)*np.sin(x**2+y)
il = lambda x: np.sqrt(1-x**2)
# il 本身已经是函数对象，可直接写成 il。
# -il 不行，因为 Python 会把函数对象取负号，会报错。
# 需要“函数返回值的相反数”时，就必须用 lambda x: -il(x) 或 lambda x: -il(x) 的具名函数。
# lambda x: -il(x) 的具名函数，即：
# def lower_bound(x):
#     return -il(x)
print(f"第二个二重积分 = {dblquad(f2, -1, 1, lambda x: -il(x), il)}")
```

计算$\iiint\limits_{\Omega} z\sqrt{x^2+y^2+1} dxdydz$，其中 $\Omega$ 为柱面 $x^2+y^2-2x=0$ 与 $z=0, z=6$ 两平面所围成的空间区域

先将三重积分化成累次积分$$I\,=\,\int_{0}^{2}\,dx\int_{-\sqrt{2x-x^2}}^{\sqrt{2x-x^2}}\,dy\int_{0}^{6}\,z\sqrt{x^2+y^2+1}\,dz$$

```python
import numpy as np
from scipy.integrate import tplquad
# 被积函数 func 的格式为 func(z, y, x),注意顺序，必须先对z积分，再对y积分，最后才对x积分，否则结果不一样（结果错误）
f = lambda z, y, x: z*np.sqrt(x**2+y**2+1)
ylu = lambda x: np.sqrt(2*x-x**2)
print(f"三次积分的结果为：{tplquad(f, 0, 2, lambda x: -ylu(x), ylu, 0, 6)}")
```
